```python
import numpy as np
import matplotlib.pyplot as plt
import os

results_path = os.path.join(os.getcwd(), 'results', 'Ant_custom', 'multi', 'full_f.npy')
```

```python
log_fitness = np.load(results_path, allow_pickle=True)
print(log_fitness.shape)

mean_fitness_0 = np.mean(log_fitness[:, :, 0], axis=1)
std_fitness_0 = np.std(log_fitness[:, :, 0], axis=1)
mean_fitness_1 = np.mean(log_fitness[:, :, 1], axis=1)
std_fitness_1 = np.std(log_fitness[:, :, 1], axis=1)
```

```python
fig, ax = plt.subplots(1,2, figsize=(12, 6))
ax[0].plot(mean_fitness_0, label='Fitness velocity')
ax[0].fill_between(range(len(mean_fitness_0)), mean_fitness_0 - std_fitness_0, mean_fitness_0 + std_fitness_0, alpha=0.2)
ax[1].plot(mean_fitness_1, label='Fitness control')
ax[1].fill_between(range(len(mean_fitness_1)), mean_fitness_1 - std_fitness_1, mean_fitness_1 + std_fitness_1, alpha=0.2)
ax[0].set_xlabel('Generation')
ax[1].set_xlabel('Generation')
ax[0].set_title('Fitness velocity')
ax[1].set_title('Fitness volume')
```

```python
fig, ax = plt.subplots(1,2, figsize=(12, 6))
ax[0].plot(log_fitness[:,:,0], label='Fitness')
ax[1].plot(log_fitness[:,:,1], label='Fitness')
ax[0].set_title('Fitness velocity')
ax[1].set_title('Fitness velume')
```
